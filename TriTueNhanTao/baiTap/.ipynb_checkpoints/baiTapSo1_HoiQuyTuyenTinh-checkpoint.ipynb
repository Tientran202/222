```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# Tạo 10 điểm dữ liệu ngẫu nhiên
np.random.seed(0)
x = np.random.rand(10, 1)
y = 2*x + 1 + np.random.randn(10, 1)*0.1
```

```python
# Vẽ chúng lên trục toạ độ
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Tính w0, w1 và R-squared dùng thư viện Sklearn
reg = LinearRegression().fit(x.reshape(-1,1), y)
y_pred = reg.intercept_ + reg.coef_[0]*x
r2Score = r2_score(y, y_pred)
print('w1:', reg.coef_[0])
print('w0:', reg.intercept_)
print('R2 score:', r2Score)
```

```python
# Tính w0 và w1 với công thức và in ra
x_mean = np.mean(x)
y_mean = np.mean(y)
w1 = np.sum((x-x_mean)*(y-y_mean))/np.sum((x-x_mean)**2)
w0 = y_mean - w1*x_mean
print('w1:', w1)
print('w0:', w0)
```

```python
# Vẽ đường thẳng với w0 và w1
plt.plot(x, y_pred, color = "b")
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Tính R-squared với công thức và in ra.
ss_res = np.sum((y - y_pred)**2)
ss_tot = np.sum((y - y_mean)**2)
r2 = 1 - ss_res/ss_tot
print("R-squared:", r2)
```

```python
# Tạo thêm 1 điểm dữ liệu cách rất xa 10 điểm dữ liệu ban đầu
x_new = np.append(x, 2)
y_new = np.append(y, 2)
```

```python
#tính lại w0 và w1 (công thức) và in ra.
x_mean = np.mean(x_new)
y_mean = np.mean(y_new)
w1 = np.sum((x_new-x_mean)*(y_new-y_mean))/np.sum((x_new-x_mean)**2)
w0 = y_mean - w1*x_mean
print('w1:', w1)
print('w0:', w0)
```

```python
#Vẽ đường thẳng mới với w0 và w1 mới.
y_pred = w0 + w1*x_new
plt.plot(x_new, y_pred, color = "b")
plt.scatter(x_new, y_new)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Tính lại R-squared và in ra.
r2Score = r2_score(y_new, y_pred)
print('R2 score:', r2Score)
```
